```python
import numpy as np
import matplotlib.pyplot as plt

n = 10; # Number of random samples generated
x = np.random.uniform(0., 1., (2,n)); # First row = Height of lower end; Second row = Angle/pi;
h = x[0,:]+np.sin(np.pi*x[1,:]); # Height of higher end
N_Bar = sum(h>=1); # Number of random samples hit the upper bar at y=1
Estimated_pi = 2*n/N_Bar # Estimate pi
print(Estimated_pi)

fig = plt.figure()
ax1 = plt.subplot2grid((2,2), (0,0), colspan=2)
ax2 = plt.subplot2grid((2,2), (1,0))
ax3 = plt.subplot2grid((2,2), (1,1))
fig.tight_layout()

for i in range(n):
    temp = np.random.normal()
    plot_x = [temp, temp+np.cos(np.pi*x[1,i])];
    plot_y = [x[0,i], x[0,i]+np.sin(np.pi*x[1,i])];
    if h[i]>=1:
        ax1.plot(plot_x, plot_y, color='red')
    else:
        ax1.plot(plot_x, plot_y, color='blue')
ax1.set_aspect(1.0)

ax2.plot(np.arange(1,n+1), np.pi*np.ones((n,))) 
ax2.plot(np.arange(1,n+1), 4*np.cumsum(h>=1)/np.arange(1,n+1))
ax2.grid(True)

n = 100; # Number of darts for each estimate
m = 100; # Number of estimates computed using n dart
x = np.random.uniform(0., 1., (2,n,m)); # First row = Height of lower end; Second row = Angle/pi;
h = x[0,:,:]+np.sin(np.pi*x[1,:,:]); # Height of higher end
N_Bar = np.sum(h>=1, axis=0); # Number of random samples hit the upper bar at y=1
Estimated_pi = 2*n/N_Bar; # Estimate pi
ax3.hist(Estimated_pi)

plt.show()
```
